```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Gastos y costos 2020.xlsx')
df
```

```python
df.info()
```

```python
df["FOLIO"]= df["FOLIO"].fillna("-")
```

```python
df["GASTO"]= df["GASTO"].fillna(0)
```

```python
df["TC"]= df["TC"].fillna(1)
```

```python
df
```

```python
valores_nulos=df.isnull().sum()

print(valores_nulos)
```

```python
df["IMPORTE"]=df["IMPORTE"].fillna(0)
```

```python
df["IMPORTE"]=df["IMPORTE"].replace("-",0)
```

```python
df["IVA"]=df["IVA"].replace("-",0)
```

```python
df["IVA"]=df["IVA"].fillna(0)
```

```python
df["TIPO"] = df["IVA"].fillna("0")
```

```python
df = df.drop('POLIZA', axis=1)
```

```python
valores_nulos=df.isnull().sum()

print(valores_nulos)
```

```python
df.to_csv('gastos_costos_2020_sin_nulos.csv')
```
