```python
#Verificar la cantidad de valores nulos por columna
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Total de valores nulos en todo el dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
df2=df.copy()
```

```python
#Sustitución de valores nulos de la columna "subtotal_ventas_alimentos_bebidas"
#por el promedio

df2["subtotal_ventas_alimentos_bebidas"]=df2["subtotal_ventas_alimentos_bebidas"].fillna(round(df2["subtotal_ventas_alimentos_bebidas"].mean(),1))  
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de los valores nulos de la columna "tarjetas_debito" por la mediana
#de la misma

df2["tarjetas_debito"]=df2["tarjetas_debito"].fillna(round(df2["tarjetas_debito"].median(),1))
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "tarjetas_credito" por el valor 20

df2["tarjetas_credito"]=df2["tarjetas_credito"].fillna(20)
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "salon_ventas" por el string "Datos faltantes"

df2["salon_ventas"]=df2["salon_ventas"].fillna("Datos faltantes")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "otros_medios" por el valor anterior

df2["otros_medios"]=df2["otros_medios"].fillna(method="ffill")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "bebidas" por el valor posterior o que sigue

df2["bebidas"]=df2["bebidas"].fillna(method="bfill")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
##Sustitución de valores nulos de la columna "almacen" por el valor posterior o que sigue
df2["almacen"]=df2["almacen"].fillna(method="bfill")
df2 
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "panaderia" por el promedio
df2["panaderia"]=df2["panaderia"].fillna(round(df2["panaderia"].mean(),1))  
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columa "lacteos" por la mediana
df2["lacteos"]=df2["lacteos"].fillna(round(df2["lacteos"].median(),1))  
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "carnes" por el dígito 30
df2["carnes"]=df2["carnes"].fillna(30)
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
# Sustitución de valores nulos de la columna "verduleria_fruteria" por la expresión "00 00"
df2["verduleria_fruteria"]=df2["verduleria_fruteria"].fillna("00 00")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "alimentos_preparados_rotiseria" por el valor anterior
df2["alimentos_preparados_rotiseria"]=df2["alimentos_preparados_rotiseria"].fillna(method="ffill")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "indumentaria_calzado_textiles_hogar" por el valor posterior
df2["indumentaria_calzado_textiles_hogar"]=df2["indumentaria_calzado_textiles_hogar"].fillna(method="bfill")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```

```python
#Sustitución de valores nulos de la columna "electronicos_articulos_hogar" por el valor posterior
df2["electronicos_articulos_hogar"]=df2["electronicos_articulos_hogar"].fillna(method="bfill")
df2
```

```python
#Comprobación de valores nulos existentes
valores_nulos=df2.isnull().sum()
valores_nulos
```
